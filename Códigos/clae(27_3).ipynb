```python
#simulo X
import numpy as np
import matplotlib.pyplot as plt

N = 1000 #num de muestras

X1 = np.random.rayleigh(scale = 3, size = N)
X2 = np.random.rayleigh(scale = 2, size = N)

X = np.array([X1, X2])
```

$
V = BX
$

```python
#simulo V


B = np.array(
    [[0.6, -0.2],
    [0.4, 0.7]]
)

V = np.dot(B, X)
```

```python
H = np.array(
    [[0.6, -0.2],
    [0.4, 0.2]]
)

U = np.dot(H, X)
```

```python
#Graficando ando

plt.scatter(X1, X2)
plt.grid(True)
plt.xlim(-2, 12)
plt.ylim(0, 14)
plt.show()
rho = np.corrcoef(X) #el elemento ij tiene el coef de corr de xi contra xj. La matriz es simétrica
r = rho[0][1]
print(r)
```

```python
plt.scatter(V[0, :], V[1, :])
plt.grid(True)
plt.xlim(-2, 12)
plt.ylim(0, 14)
plt.show()
rho = np.corrcoef(V) #el elemento ij tiene el coef de corr de xi contra xj. La matriz es simétrica
r = rho[0][1]
print(r)
```

```python
#Primero estimo la media

M = X #matriz de muestas
mu1 = np.mean(M[0, :])
mu2 = np.mean(M[1, :]) #estimaciones de las medias

mu = np.array([mu1, mu2]) #vector con estimaciones

#sale un for?
C_x = np.zeros(shape = (2, 2))

for i in range(N):
    vec = M[: , i] - mu
    vec_columna = vec.reshape(-1,1)
    C_x = C_x + vec_columna@vec_columna.T
C_x = 1/(N-1) * C_x
print(C_x)
```

```python
#hago al función para comparar
C_comparar = np.cov(M)
print(C_comparar)
```

```python
X1 = np.random.uniform(low = 0, high = 2, size = N)
X2 = np.random.uniform(low = 0, high = 3, size = N)

X = np.array([X1, X2])

th = 0 #ángulo thita de la rotación

R = np.array([[np.cos(th), (-1)*np.sin(th)],[np.sin(th), np.cos(th)]])

Y = np.dot(R, X)

plt.scatter(X1, X2) #acá van las variables que vas a scatterear. Ahre, qué era ese verbo
plt.grid(True)
plt.xlim(-1 , 4) #Tus límites
plt.ylim(-1, 4)
plt.show()

plt.scatter(Y[0, :], Y[1, :]) #acá van las variables que vas a scatterear. Ahre, qué era ese verbo
plt.grid(True)
plt.xlim(-1 , 4) #Tus límites
plt.ylim(-1, 4)
plt.show()
```
