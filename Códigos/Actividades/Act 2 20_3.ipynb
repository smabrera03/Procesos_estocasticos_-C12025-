Sea x una variable aleatoria exponencial, X ∼ Exp(λ), de parámetro λ = 0.5
1. Genere N = 10.000
 muestras de X (usando el método de transformación

Invirtiendo se llega a que:
y0 es una realización de la normal

x0 es la realización de la exponencial que busco. La relación es:

$
x_0 = \frac{-1}{\lambda}\ln(1-y_0)
$


```python
import numpy as np
import matplotlib.pyplot as plt

N = 10000
#lambda = 0.5 no se puede, lambda es una palabra reservada
l = 0.5

y = np.random.rand(N) #simulo la VA uniforme N veces
x = (-1/l)*np.log(1-y)
```

```python
#habrá salido bien? Grafiquemos:
bins = 30

plt.hist(x, bins, density = True)

t = np.linspace(start = 0, stop =  max(x), num = 500)
pdf = l*np.exp(-l*t)

plt.plot(t, pdf)
plt.grid(True)

plt.show()
```

Siguiente punto: Estime la media y la varianza muestrales. ¿Cómo?

La función np.mean(x) hace el promedio.

Otra idea: aproximo la integral por el histograma:

$
\int_{0}^{\infty} x.f(x) \,dx \approx \sum_{\forall x_0} x_0 . h(x_0)
$

Donde $h(x_0)$ es la altura del histogramam en ese punto

¿Son lo mismo estas dos formas?¿Por qué?

En cuanto a la varianza:

$
\hat{\sigma}^2 = \frac{1}{N} \sum_{\forall x_0} (x_0 - \hat{\mu})^2
$
 


```python
mu = np.mean(x)
var = np.var(x)
std = np.sqrt(var)
```

```python
#en el grafico se ven así:

#habrá salido bien? Grafiquemos:
bins = 30

plt.hist(x, bins, density = True)

t = np.linspace(start = 0, stop =  max(x), num = 500)
pdf = l*np.exp(-l*t)

plt.plot(t, pdf)
plt.xticks([mu, mu - std, mu + std], ['$\mu$', '$\mu - \sigma$', '$\mu + \sigma$'])

plt.grid(True)
plt.show()
```
