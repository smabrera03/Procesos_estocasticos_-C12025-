```python
import numpy as np
import scipy.signal as sg
import matplotlib.pyplot as plt

a0 = 1
b0 = 1
b1 = 0.8
b2 = 0.4
b = [b0, b1, b2]
N = 1000

X = np.random.randn(N)
Y = sg.lfilter(b, a0, X)
```

```python
plt.plot(X, label = 'X(n)')
plt.plot(Y, label = 'Y(n)')

#plt.xlim(0, 100)

plt.legend()
plt.grid(True)
plt.show()
```

```python
#Correlaciones

Rx = 1/N * np.correlate(X, X, mode = 'full')
Ry = 1/N * np.correlate(Y, Y, mode = 'full')

tau = np.arange(start = -(N-1), stop = N)

plt.plot(tau, Rx, label = f'$R_X(\\tau)$')
plt.plot(tau, Ry, label = f'$R_Y(\\tau)$')

plt.xlim(-5, 5)

plt.legend()
plt.grid(True)
plt.show()
```

```python
#PSD´s estimadas
PSD_X = (1/N) * np.abs(np.fft.fft(X, n = N))**2
PSD_Y = (1/N) * np.abs(np.fft.fft(Y, n = N))**2

w = np.linspace(start = 0, stop = 2*np.pi, num = len(PSD_X))

#PSD´s teóricas
S_X = np.ones_like(w) #Cooorta

#Para S_Y lo voy a hacer a lo bruto, calculando valores de H para algunos omega
H = b0 * np.ones_like(w) + b1 * np.exp(-w * 1j) + b2 * np.exp(-2*w * 1j)

S_Y = np.abs(H)**2

plt.plot(w, PSD_X, label = f'$S_X(\\omega) estimado$')
plt.plot(w, S_X, label = f'$S_X (\\omega) teórica $')

plt.xlim(0, 2*np.pi)

plt.legend()
plt.grid(True)
plt.show()


plt.plot(w, PSD_Y, label = f'$S_Y(\\omega) estimado$')
plt.plot(w, S_Y, label = f'$S_Y(\\omega) teórica$')

plt.xlim(0, 2*np.pi)

plt.legend()
plt.grid(True)
plt.show()
```
