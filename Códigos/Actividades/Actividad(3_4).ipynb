```python
import numpy as np
import matplotlib.pyplot as plt

N = 5000

Z = np.random.randn(2, N) #2 filas, N columnas
Z1 = Z[0, :]
Z2 = Z[1, :]

bins = 30
plt.hist(Z1, bins, density=True)
plt.grid(True)
plt.show()

plt.hist(Z2, bins, density=True)
plt.grid(True)
plt.show()

plt.scatter(Z1, Z2)
plt.grid(True)
```

```python
C_Y = np.array([[0.7, 0.9], 
                 [0.9, 1.75]])

mu_Y = np.array([0.8, 1.9])
a_val, P = np.linalg.eig(C_Y)
#nota: P[:, i] es el i-esimo autovector asociado al i-esimo autovalor, que es a_val[i]
Lambda = np.diag(a_val) #matriz con a_vals en la diagonal
Lambda = np.sqrt(Lambda) #Le tomo la raz

print(P, Lambda)

A = np.dot(P, Lambda)

print(A)
```

```python
Y = A@Z + mu_Y[:, np.newaxis]

Y1 = Y[0,:]
Y2 = Y[1, :]

plt.hist(Y1, bins, density=True)
plt.grid(True)
plt.show()

plt.hist(Y2, bins, density=True)
plt.grid(True)
plt.show()

plt.scatter(Y1, Y2)
plt.grid(True)

```
