```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Parmetros de la gaussiana
mu = np.array([1, -2])  # media
Sigma = np.array([[2, 1], [1, 1]])  # matriz de covarianza

# Crear una malla de puntos para evaluar la densidad
x, y = np.mgrid[-3:5:.01, -5:1:.01] 
#esto es: en x, de menos 5 a 5 con saltos de 0.1
pos = np.dstack((x, y))  # grid de coordenadas (N, N, 2)

# Evaluar la densidad en cada punto
rv = multivariate_normal(mean=mu, cov=Sigma) #rv es un objeto que representa la normal. rv de random vecotr
z = rv.pdf(pos) #se evalua la pdf del rv


#grafico autovectores

avals, avecs = np.linalg.eigh(Sigma)

plt.quiver(*mu, *avecs[0], scale = 1/np.sqrt(avals[0]), angles='xy', scale_units='xy', color='red', width=0.008)
print(*mu)
print(avecs[0], avecs[1])
plt.quiver(*mu, *avecs[1], scale = 1/np.sqrt(avals[1]), scale_units = 'xy', color = 'blue')
#Nota: quiver divide al vector a dibujar por scale.
#Nota: la funcin quiver dibuja vectores. Los dos primeros parmetros son el punto de origen y los dos segundos
#son la direccin
# Graficar curvas de nivel
plt.contour(x, y, z, levels=5, cmap="viridis")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Curvas de nivel de la densidad')
plt.axis('equal')
plt.grid(True)
plt.show()
```
