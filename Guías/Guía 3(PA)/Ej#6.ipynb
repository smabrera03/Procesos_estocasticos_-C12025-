$X(n)$ depdende del valor que haya tenido $X(n-1)$

$P[X(0) = 1] = p$ ==> Semilla?

```python
import numpy as np
import matplotlib.pyplot as plt
p = 0.5 #la probailidad de que X arranque en 1. Lo eleg xq s

l1 =  0.5
l2 = 0.1 #lambda 1 y 2

def simular_realizacion(N):
    X_0 = np.random.binomial(n = 1, p = p, size = (1,)) #simulo el primer elemento de mi proceso
    for i in range(1, N):
        if(X_0[i-1] == 0):
            B1 = np.random.binomial(n = 1, p = l1)
            X_0 = np.append(X_0, B1) #Si el anterior es 0, simulo un b1 y lo agrego
        if(X_0[i-1] == 1):
            B2 = np.random.binomial(n = 1, p = l2)
            X_0 = np.append(X_0, B2)#si el anterior es 1, simulo un b2 y lo agrego
    return X_0

def simular_realizaciones(R, N):
    X = np.zeros((R, N))
    for i in range(R):
        X[i, :] = simular_realizacion(N)
    return X

X_0 = simular_realizaciones(R = 5, N = 200)
print(X_0, X_0.shape)
```

```python
N = 200
R = 5

n = np.arange(N)
X_0 = simular_realizaciones(R = R, N = N)
for i in range(1):
    plt.step(n, X_0[i], where = 'post')

mu_obt = np.mean(X_0, axis = 0)
plt.step(n, mu_obt)

plt.grid()
plt.show()
```
