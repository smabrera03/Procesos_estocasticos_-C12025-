### m = 1

$
X(n) + a_1\cdot X(n - 1) = W(n)
$

$
X(n) = W(n) - a_1\cdot X(n - 1)
$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 1 #orden de mi sistema AR
RN = np.array([[2]])
rN = np.array([0.8])

aN = - np.linalg.inv(RN) @ rN #NOTA: el @ es el producto matricial

Var = RN[0, 0] + np.transpose(rN) @ aN
sigma = np.sqrt(Var)
#genero una realización de W(n) de 20 elementos y con ella una realización de X(n)
N = 20
Realizaciones  = 100
W = sigma * np.random.randn(Realizaciones, N) #realizaciones en las filas, eje de tiempos en las columnas
X = np.zeros_like(W)

for j in range(Realizaciones):
    X[j, 0] = W[j, 0]
    for i in range(1, N):
        X[j, i] = W[j, i] - aN[0]*X[j, i-1]

RX = np.zeros((Realizaciones, 2*N - 1))
for j in range(Realizaciones):
    RX[j] = (1/N) * np.correlate(X[j], X[j], mode = 'full')
RX = np.mean(RX, axis = 0)
k = np.arange(start = -(N-1), stop = N)
plt.stem(k, RX)

plt.xlim(-4, 4)

plt.grid(True)
plt.show()
```

### m = 2

$
X(n) + a_1\cdot X(n - 1) + a_2\cdot X(n-2) = W(n)
$

$
X(n) = W(n) - a_1\cdot X(n - 1) - a_2\cdot X(n-2)
$

```python
m = 2 #orden de mi sistema AR
RN = np.array([[2, 0.8], [0.8, 2]])
rN = np.array([0.8, 0.82])

aN = - np.linalg.inv(RN) @ rN #NOTA: el @ es el producto matricial

Var = RN[0, 0] + np.transpose(rN) @ aN
sigma = np.sqrt(Var)
#genero una realización de W(n) de 20 elementos y con ella una realización de X(n)
N = 50
Realizaciones = 1000
W = sigma * np.random.randn(Realizaciones, N) #realizaciones en las filas, eje de tiempos en las columnas
X = np.zeros_like(W)

for j in range(Realizaciones):
    X[j, 0] = W[j, 0]
    X[j, 1] = W[j, 1] - aN[0]*X[j, 0]
    for i in range(2, N):
        X[j, i] = W[j, i] - aN[0]*X[j, i-1] - aN[1]*X[j, i-2]  

RX = np.zeros((Realizaciones, 2*N - 1))
for j in range(Realizaciones):
    RX[j] = (1/N) * np.correlate(X[j], X[j], mode = 'full')
RX = np.mean(RX, axis = 0)
k = np.arange(start = -(N-1), stop = N)
plt.stem(k, RX)

plt.xlim(-4, 4)

plt.grid(True)
plt.show()
```

### m = 3

$
X(n) + a_1\cdot X(n - 1) + a_2\cdot X(n-2) + a_3 \cdot X(n-3)= W(n)
$

$
X(n) = W(n) - a_1\cdot X(n - 1) - a_2\cdot X(n-2) - a_3 \cdot X(n-3)
$

```python
m = 2 #orden de mi sistema AR
RN = np.array([[2, 0.8, 0.82], 
               [0.8, 2, 0.8], 
               [0.82, 0.8, 2]])
rN = np.array([0.8, 0.82, 0.728])

aN = - np.linalg.inv(RN) @ rN #NOTA: el @ es el producto matricial

Var = RN[0, 0] + np.transpose(rN) @ aN
sigma = np.sqrt(Var)
#genero una realización de W(n) de 20 elementos y con ella una realización de X(n)
N = 50
Realizaciones = 1000
W = sigma * np.random.randn(Realizaciones, N) #realizaciones en las filas, eje de tiempos en las columnas
X = np.zeros_like(W)

for j in range(Realizaciones):
    X[j, 0] = W[j, 0]
    X[j, 1] = W[j, 1] - aN[0]*X[j, 0]
    X[j, 2] = W[j, 2] - aN[0]*X[j, 1] - aN[1]*X[j, 0]
    for i in range(3, N):
        X[j, i] = W[j, i] - aN[0]*X[j, i-1] - aN[1]*X[j, i-2] - aN[2]*X[j, i-3] 

RX = np.zeros((Realizaciones, 2*N - 1))
for j in range(Realizaciones):
    RX[j] = (1/N) * np.correlate(X[j], X[j], mode = 'full')
RX = np.mean(RX, axis = 0)
k = np.arange(start = -(N-1), stop = N)
plt.stem(k, RX)

plt.xlim(-4, 4)

plt.grid(True)
plt.show()
```
